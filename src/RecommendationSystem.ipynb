```python
import pandas as pd
import numpy as np

movie = pd.read_csv('../dataset/movie.csv')
rating = pd.read_csv('../dataset/rating.csv')
data  = pd.read_csv('../dataset/processed_data.csv')
print(data.shape,data['movieId'].nunique(),data['userId'].nunique())
USER_ID = 7
user_df = data[data['userId']==USER_ID]
print(user_df.shape,user_df['movieId'].nunique())
user_df.head()
```

```python
movies_avg_rating = data.groupby('movieId').agg('mean')
```

```python
movies_avg_rating
```

```python
movies_watched = user_df['movieId'].unique()
genres =[]
for i, row in user_df.iterrows():
    row_genres = row['genres'].split('|')
    genres.extend(row_genres)
    genres = list(set(genres))
```

```python
movies_watched,genres
```

```python
# getting users who have watched 60% of movies watched by user 1
similar_users = data[data['movieId'].isin(user_df['movieId'])].groupby(['userId'])['movieId'].count().reset_index().sort_values(by='movieId',ascending=False)
similar_users=similar_users[similar_users['movieId']>=0.6*user_df['movieId'].nunique()]
similar_users = similar_users['userId'].tolist()
print(sorted(similar_users)),print(len(similar_users))
```

```python
corr = data[data['userId'].isin(similar_users)]
mat =corr.pivot_table(index='userId',columns='movieId',values='rating')
```

```python
corr_df = pd.DataFrame(mat.T.corr().unstack().sort_values(), columns=['corr'])
corr_df.index.names = ['user_id_1', 'user_id_2']

corr_df.reset_index(inplace=True)
corr_df.describe()
```

```python
top_users = corr_df[(corr_df["user_id_1"] == USER_ID) & (corr_df["corr"] >= corr_df['corr'].mean())][["user_id_2", "corr"]].reset_index(drop=True).sort_values(by="corr", ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
top_users
```

```python
top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')
top_users_ratings = top_users_ratings[top_users_ratings["userId"] != USER_ID]
```

```python
top_users_ratings
```

```python
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df.sort_values(by='weighted_rating', ascending=False)
```

```python
movies_to_recommend = recommendation_df.sort_values(by='weighted_rating', ascending=False)
print(movies_to_recommend)
movies_to_recommend.merge(movie[['movieId','title']]).head(10)
```

# Recommendation based on the item


```python
movie_id = rating[(rating["userId"] == USER_ID) & (rating["rating"] == 5.0)].sort_values(by="timestamp", ascending=False)["movieId"][0:1].values[0]
recent_liked_movie = movie[movie["movieId"] == movie_id]["movieId"].values[0]
movie_df = data[data["movieId"] == movie_id].sort_values(by='userId')
temp = mat[recent_liked_movie]
res = pd.DataFrame(mat.corrwith(temp).sort_values(ascending=False), columns=["corr"])

# movie_df
```

```python
temp = res.merge(movies_avg_rating, on='movieId').drop(['userId'],axis=1)
temp['score']= temp['corr']*temp['rating']
temp.sort_values(by='score', ascending=False).head(10)
```

```python
recommedations = movies_to_recommend.merge(temp,on='movieId').dropna()
recommedations['final_score'] = (recommedations['score']+recommedations['weighted_rating'])/2
```

```python
recommedations.sort_values(by='final_score', ascending=False, inplace=True)
```
